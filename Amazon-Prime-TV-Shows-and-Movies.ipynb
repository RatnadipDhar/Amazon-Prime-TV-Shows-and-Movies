## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# 1. Remove duplicate rows
df = df.drop_duplicates()

# 2. Handle missing values
# Drop columns with too many missing values
df = df.dropna(axis=1, thresh=len(df)*0.5)# keep columns with at least 50% data
# Fill missing values for important columns
df['imdb_score'] = df['imdb_score'].fillna(df['imdb_score'].mean())
df['tmdb_score'] = df['tmdb_score'].fillna(df['tmdb_score'].mean())
df['runtime'] = df['runtime'].fillna(df['runtime'].median())
df['genres'] = df['genres'].fillna('Unknown')

# 3. Convert data types
df['release_year'] = df['release_year'].astype(int)

# 4. Remove unnecessary columns
df = df.drop(columns=['description'], errors='ignore')

# 5. Strip extra spaces for string columns
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].str.strip()

# 6. Check final dataset
print("Final Dataset Shape:", df.shape)
df.head()
````

### What all manipulations have you done and insights you found?

The dataset was cleaned and organized so that it could be used effectively for analysis. Missing values were handled carefully by filling them with suitable values, and duplicate entries were removed to avoid incorrect results. Unnecessary columns were dropped to keep the dataset simple and relevant. Both datasets were then combined using a common column, which helped in bringing together all the important information in one place. Small corrections like fixing data types and cleaning text values were also done to make the data more consistent.
After these changes, the dataset became much more structured and easier to work with. It was noticed that some information, especially ratings and runtime, was missing in places, and there were repeated entries that needed cleaning. The combined dataset now provides a clearer view of movies and shows along with their related details, making it suitable for further analysis and building models.
